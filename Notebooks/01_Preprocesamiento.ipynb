```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Valor total de mercado de la oferta circulante de todas las criptomonedas,
#incluidas las stablecoins y los tokens.
df_marketcap = pd.read_csv("../Data/Raw/mc.csv", parse_dates = ['DateTime'], sep=";")
df_marketcap
```

```python
df_marketcap.info()
```

```python
print(df_marketcap['DateTime'].min())
print(df_marketcap['DateTime'].max())
```

```python
df_marketcap.set_index('DateTime', inplace=True)
df_marketcap.head()
```

```python
df_marketcap['Market cap'] = df_marketcap['Market cap'].str.replace(',', '.').astype(float)
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_marketcap);
```

```python
#Volumen total de operaciones al contado en 24 horas de todas las criptomonedas,
#incluidas las stablecoins y los tokens.
df_volume = pd.read_csv("../Data/Raw/volume24h.csv", parse_dates = ['DateTime'], sep=";")
df_volume
```

```python
df_volume.set_index('DateTime', inplace=True)
df_volume['Volume (24h)'] = df_volume['Volume (24h)'].str.replace(',', '.').astype(float)
```

```python
#La capitalizaciÃ³n bursÃ¡til total de Bitcoin comparada con la
#capitalizaciÃ³n bursÃ¡til total de todas las criptomonedas combinadas.
df_dom = pd.read_csv("../Data/Raw/btc_dominance.csv", parse_dates = ['DateTime'], sep=";")
df_dom
```

```python
df_dom.set_index('DateTime', inplace=True)
df_dom['BTC'] = df_dom['BTC'].str.replace(',', '.').astype(float)
df_dom['ETH'] = df_dom['ETH'].str.replace(',', '.').astype(float)
df_dom['USDT'] = df_dom['USDT'].str.replace(',', '.').astype(float)
df_dom['BNB'] = df_dom['BNB'].str.replace(',', '.').astype(float)
df_dom['XRP'] = df_dom['XRP'].str.replace(',', '.').astype(float)
df_dom['Otros'] = df_dom['Otros'].str.replace(',', '.').astype(float)
```

```python
df_dom.info()
```

```python
df_dom["Total"] = df_dom.sum(axis=1)
df_dom
```

```python
df_dompercentage = df_dom[['BTC', 'ETH', 'USDT', 'BNB', 'XRP', 'Otros']].div(df_dom['Total'], axis=0) * 100
```

```python
df_dompercentage.head()
```

```python
dfs = [df_marketcap,df_volume,df_mcbtc, df_mceth, df_mcusdt, df_mcbnb, df_mcxrp, df_mcusdc, df_mcsol, df_mcada, df_mcdoge, df_mctrx]

result_df = pd.concat(dfs, axis=1)
result_df
```

```python
sns.pairplot(result_df);
```

```python
sns.heatmap(result_df.corr(), annot=True);
```

```python
result_df[['Market cap BTC', 'Volume (24h)']].plot(figsize = (16, 8))
plt.ylabel('Price (USD)')
plt.yscale('log')
plt.title('Volume of transactions')
```

```python
result_df
```

```python
result_df['MA200'] = result_df['Market cap'].rolling(window = 200).mean()
```

```python
result_df['MA200/BTC'] = result_df['Market cap BTC'].rolling(window = 200).mean()
result_df.tail()
```

```python
result_df['MA200/ETH'] = result_df['Market cap ETH'].rolling(window = 200).mean()
```

```python
result_df['MA200/USDT'] = result_df['Market cap USDT'].rolling(window = 200).mean()
```

```python
result_df['MA200/USDT']
```

```python
result_df[['Market cap BTC', 'MA200/BTC']].plot(figsize = (16, 8))
plt.ylabel('Price (USD)')
plt.title('Price & 200 Day MA')
```

```python
result_df['MM'] = result_df['Market cap BTC'] / result_df['MA200/BTC']
result_df['MM'].plot(figsize = (16, 8), grid = True, color = 'black', lw = 0.8)
plt.title('Mayer Multiple')
plt.ylabel('Price / 200MA')
```

```python
result_df['NVT'] = result_df['Market cap'] / result_df['Volume (24h)']
```

```python
result_df['NVT'].plot(figsize = (16, 8), grid = True, color = 'green', lw = 0.8)
plt.title('NVT-value')
plt.ylabel('NVT')
```

```python
result_df['NVT-smooth'] = result_df['NVT'].rolling(window = 14).mean()
result_df['NVT-smooth'].plot(figsize = (16, 8), lw = 0.8, grid = True)
plt.title('14MA NVT')
plt.ylabel('NVT')
```

```python
result_df = result_df.drop(columns="NVT")
```
